# Binance

```python
url = 'https://api.bscscan.com/api\
?module=account\
&action=tokentx\
&address=0x0C89C0407775dd89b12918B9c0aa42Bf96518820\
&page=1\
&offset=100\
&startblock=0\
&endblock=999999999\
&sort=asc\
&apikey=6ECWSUUEEJ8FJH4GHKSS7F58Z8NY45DJJ8'

r = requests.get(url)
r.status_code
data = r.json()
dfb = pd.DataFrame(data['result'])
dfb
```

```python
len(data['result'])
```

```python
dfs = []
error = 0
starttime = datetime.datetime.now()
endblock = 18900000
transaction = 10000

for block in range(0, endblock, transaction):
    url = f'https://api.etherscan.io/api\
?module=account\
&action=tokentx\
&address=0xe2fe530c047f2d85298b07d9333c05737f1435fb\
&page=1\
&offset=10000\
&startblock={block}\
&endblock={block+9999}\
&sort=asc\
&apikey=SE4Y8HN6P577WGIJSM8561XVS3279RCYAS'
    
    r = requests.get(url)
    data = r.json()
    df = pd.DataFrame(data['result'])
    length = len(data['result'])
    if length == 10000:
        error += 1
        print(f'{block}/{endblock}区块链上溢出!')
    
    dfs.append(df)
    
    endtime = datetime.datetime.now()
    duration = (endtime-starttime).seconds / 60
    lefttime = (duration / (block + 1)) * (18900000- block)
    print(f'成功获取{block}/{endblock}区块链上的数据! 预估还需要{lefttime}分钟, 数据长度为{length}, 溢出次数为{error}')
```

```python
dfs = pd.concat(dfs)
dfs
dfs.to_csv(f'{WD}tf_binance.csv', index=False)
```
