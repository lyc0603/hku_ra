```python
df = pd.read_csv(f'{WD}processed/root_ethereum_unwrapped.csv')
ethereum = ['uniswapv3_ethereum', 'uniswapv2_ethereum', 'sushiswap_ethereum']
df[ethereum] = -1
df
df.to_csv(f'{WD}insider/list/root_ethereum_list.csv', index=False)
```

```python
dexes = {
'ethereum' : ['uniswapv3_ethereum', 'uniswapv2_ethereum', 'sushiswap_ethereum'],
'binance': ['pancake_binance', 'biswap_binance', 'apeswap_binance', 'babyswap_binance', 'jetswap_binance'],
'avalanche' : ['traderjoe_avalanche', 'pangolin_avalanche'],
'fantom' : ['spookyswap_fantom', 'spiritswap_fantom'],
'polygon' : ['quickswap_polygon', 'sushiswap_polygon', 'apeswap_polygon', 'jetswap_polygon']
}

chains = ['ethereum', 'binance', 'avalanche', 'fantom', 'polygon']

chain_num = 2
dex_num = 1
check_num = 0

dex = dexes[chains[chain_num]][dex_num]

df = pd.read_csv(f'{WD}insider/list/root_{dex}_list.csv')
for idx, row in df.loc[df[dex] < 0].iterrows():
    dfw = []
    dfs = pd.read_csv(f'{WD}{dex}.csv', chunksize=10**6, iterator=True)
    n = 0
    num = 0
            
    name = row['tokenName']
    root = row['root']

    for chunk in dfs:
        if check_num == 0:
            check = (chunk['from'] == root) | (chunk['to'] == root) | (chunk['sender'] == root)
        if check_num == 1:
            check = (chunk['origin'] == root) | (chunk['recipient'] == root) | (chunk['sender'] == root)
        if check_num == 2:
            check = (chunk['sender'] == root) | (chunk['to'] == root)
            
        length = chunk.loc[check].shape[0]

        if length != 0:
            dfw.append(chunk.loc[check])

        n += 1
        num = num + length
        
        print(f'本次搜索对象为{name}, 已搜索{dex}交易所第{n*10**6}个数据, 本次内幕交易次数为{length}')

    print(f'{name}在{dex}上遍历完成!累计搜索到{num}组内幕交易.')
    if num != 0:
        dfw = pd.concat(dfw)
        dfw.to_csv(f'{WD}insider/{root}_{dex}.csv', index=False) 

    df.loc[df['root'] == root, dex] = length
    df.to_csv(f'{WD}insider/list/root_{dex}_list.csv', index=False)
```

# 多线程遍历

```python
class insider(threading.Thread):
    
    def __init__(self, chain_num, dex_num, check_num):
        threading.Thread.__init__(self)
        self.chain_num = chain_num
        self.dex_num = dex_num
        self.check_num = check_num
        
    def run(self):
        dexes = {
        'ethereum' : ['uniswapv3_ethereum', 'uniswapv2_ethereum', 'sushiswap_ethereum'],
        'binance': ['pancake_binance', 'biswap_binance', 'apeswap_binance', 'babyswap_binance', 'jetswap_binance'],
        'avalanche' : ['traderjoe_avalanche', 'pangolin_avalanche'],
        'fantom' : ['spookyswap_fantom', 'spiritswap_fantom'],
        'polygon' : ['quickswap_polygon', 'sushiswap_polygon', 'apeswap_polygon', 'jetswap_polygon']
        }

        chains = ['ethereum', 'binance', 'avalanche', 'fantom', 'polygon']
        
        dex = dexes[chains[self.chain_num]][self.dex_num]

        df = pd.read_csv(f'{WD}insider/list/root_{dex}_list.csv')
        
        for idx, row in df.loc[df[dex] < 0].iterrows():
            dfw = []
            dfs = pd.read_csv(f'{WD}{dex}.csv', chunksize=10**6, iterator=True)
            n = 0
            num = 0

            name = row['tokenName']
            root = row['root']

            for chunk in dfs:
                if self.check_num == 0:
                    check = (chunk['from'] == root) | (chunk['to'] == root) | (chunk['sender'] == root)
                if self.check_num == 1:
                    check = (chunk['origin'] == root) | (chunk['recipient'] == root) | (chunk['sender'] == root)
                if self.check_num == 2:
                    check = (chunk['sender'] == root) | (chunk['to'] == root)
                
                length = chunk.loc[check].shape[0]

                if length != 0:
                    dfw.append(chunk.loc[check])

                n += 1
                num = num + length

                # print(f'本次搜索对象为{name}, 已搜索{dex}交易所第{n*10**6}个数据, 本次内幕交易次数为{length}')

            print(f'{name}在{dex}上遍历完成!累计搜索到{num}组内幕交易.')
            if num != 0:
                dfw = pd.concat(dfw)
                dfw.to_csv(f'{WD}insider/{root}_{dex}.csv', index=False) 

            df.loc[df['root'] == root, dex] = length
            df.to_csv(f'{WD}insider/list/root_{dex}_list.csv', index=False)
            
        print(f'***{dex}遍历已完成***')
```

```python
thread1 = insider(0, 0, 1)
thread2 = insider(0, 1, 0)
thread3 = insider(0, 2, 2)

thread4 = insider(1, 0, 0)
thread5 = insider(1, 1, 0)
thread6 = insider(1, 2, 0)
thread7 = insider(1, 3, 0)
thread8 = insider(1, 4, 0)

thread9 = insider(2, 0, 2)
thread10 = insider(2, 1, 0)

thread11 = insider(3, 0, 2)
thread12 = insider(3, 1, 2)

thread13 = insider(4, 0, 0)
thread14 = insider(4, 1, 2)
thread15 = insider(4, 2, 0)
thread16 = insider(4, 3, 0)
```

```python
Threads = [
           thread1, 
           thread2, 
           thread3,
           thread4,
           thread5,
           thread6,
           thread7,
           thread8,
           thread9,
           thread10,
           thread11,
           thread12,
           thread13,
           thread14,
           thread15,
           thread16,
          ]
for t in Threads:
    t.start()
for t in Threads:
    t.join()
```

# 查看进度

```python
dexes = {
'ethereum' : ['uniswapv3_ethereum', 'uniswapv2_ethereum', 'sushiswap_ethereum'],
'binance': ['pancake_binance', 'biswap_binance', 'apeswap_binance', 'babyswap_binance', 'jetswap_binance'],
'avalanche' : ['traderjoe_avalanche', 'pangolin_avalanche'],
'fantom' : ['spookyswap_fantom', 'spiritswap_fantom'],
'polygon' : ['quickswap_polygon', 'sushiswap_polygon', 'apeswap_polygon', 'jetswap_polygon']
}

chains = ['ethereum', 'binance', 'avalanche', 'fantom', 'polygon']

done_sum = 0
total_sum = 0

for chain in chains:
    for dex in dexes[chain]:
        df = pd.read_csv(f'{WD}insider/list/root_{dex}_list.csv')
        done = df.loc[df[dex] >= 0].shape[0]
        total = df.shape[0]
        percentage = done/total
        
        done_sum += done
        total_sum += total


        print(f'{dex}已完成{done}/{total} ({percentage})')
percentage_sum = done_sum/total_sum        
print(f'\n总进度{done_sum}/{total_sum} ({percentage_sum})')
```

```python
dexes = {
'ethereum' : ['uniswapv3_ethereum', 'uniswapv2_ethereum', 'sushiswap_ethereum'],
'binance': ['pancake_binance', 'biswap_binance', 'apeswap_binance', 'babyswap_binance', 'jetswap_binance'],
'avalanche' : ['traderjoe_avalanche', 'pangolin_avalanche'],
'fantom' : ['spookyswap_fantom', 'spiritswap_fantom'],
'polygon' : ['quickswap_polygon', 'sushiswap_polygon', 'apeswap_polygon', 'jetswap_polygon']
}

chains = ['ethereum', 'binance', 'avalanche', 'fantom', 'polygon']

participants = []
insiders = []

for chain in chains:
    for dex in dexes[chain]:
        df = pd.read_csv(f'{WD}insider/list/root_{dex}_list.csv')
        
        participants.append(df.loc[:, 'root'].reset_index().drop(columns='index'))
        insiders.append(df.loc[df[dex] > 0, 'root'].reset_index().drop(columns='index'))

participants = pd.concat(participants)
insiders = pd.concat(insiders)

participant_num = participants.drop_duplicates(subset='root').shape[0]
insider_num = insiders.drop_duplicates(subset='root').shape[0]

print(f'目前为止可能涉及到内幕交易的钱包有{insider_num}/{participant_num} ({insider_num/participant_num})')
```
