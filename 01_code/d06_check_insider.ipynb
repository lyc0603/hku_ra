# Check List

```python
dexes = {
'ethereum' : ['uniswapv3_ethereum', 'uniswapv2_ethereum', 'sushiswap_ethereum'],
'binance': ['pancake_binance', 'biswap_binance', 'apeswap_binance', 'babyswap_binance', 'jetswap_binance'],
'avalanche' : ['traderjoe_avalanche', 'pangolin_avalanche'],
'fantom' : ['spookyswap_fantom', 'spiritswap_fantom'],
'polygon' : ['quickswap_polygon', 'sushiswap_polygon', 'apeswap_polygon', 'jetswap_polygon']
}

chains = ['ethereum', 'binance', 'avalanche', 'fantom', 'polygon']

done_sum = 0
total_sum = 0
no_sum = 0

for chain in chains:
    for dex in dexes[chain]:
        df = pd.read_csv(f'{WD}insider/list/root_{dex}_list.csv').drop_duplicates('root', keep='last')
        
        done = df.loc[df[dex] >= 0].shape[0]
        insider = df.loc[df[dex] >0].shape[0]
        total = df.shape[0]
        no = df.loc[df[dex] >0, dex].sum()
        percentage = done/total
        insider_percentage = insider/done
        
        done_sum += done
        total_sum += total
        no_sum += no


        print(f'{dex}已完成{done}/{total} ({percentage}), 发现潜在内幕交易者{insider}/{done} ({insider_percentage})')
        df.sort_values(f'{dex}', ascending=False)
        
percentage_sum = done_sum/total_sum        
print(f'\n总进度{done_sum}/{total_sum} ({percentage_sum})')
print(f'\n总内幕交易{no_sum}')
```

# Panel

```python
dexes = {
'ethereum' : ['uniswapv3_ethereum', 'uniswapv2_ethereum', 'sushiswap_ethereum'],
'binance': ['pancake_binance', 'biswap_binance', 'apeswap_binance', 'babyswap_binance', 'jetswap_binance'],
'avalanche' : ['traderjoe_avalanche', 'pangolin_avalanche'],
'fantom' : ['spookyswap_fantom', 'spiritswap_fantom'],
'polygon' : ['quickswap_polygon', 'sushiswap_polygon', 'apeswap_polygon', 'jetswap_polygon']
}

chains = ['ethereum', 'binance', 'avalanche', 'fantom', 'polygon']
error = ['pancake_binance', 'quickswap_polygon', 'spookyswap_fantom', 'sushiswap_polygon', 'uniswapv2_ethereum']

dfs = []
info_insider = ['root', 'tokenName', 'tokenSymbol', 'contractAddress']

for chain in chains:
    for dex in dexes[chain]:
        lst = pd.read_csv(f'{WD}insider/list/root_{dex}_list.csv').drop_duplicates('root', keep='last')
        for idx, row in lst.loc[lst[dex] > 0].iterrows():
            root = row['root']
            
            print(f'{root}_{dex}.csv')
            df = pd.read_csv(f'{WD}insider/{root}_{dex}.csv')
            
            lst.loc[lst['root'] == root, dex] = df.shape[0]
            
            # if df.shape[0] != row[dex]: print(f'出现错误！{df.shape[0]}/{row[dex]}')
            
            if dex in error: df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')  
            
            for info in info_insider:
                df[f'insider_{info}'] = row[info]
            
            df['dex'] = dex
            df['chain'] = chain
            
            df.rename(columns={'token0':'token0_name', 'token1':'token1_name'}, inplace=True)
            if 'amount0In' in df.keys():
                for idx, row in df.iterrows():
                    if row['amount0In'] == 0:
                        df.loc[df['id'] == row['id'], 'amount0'] = row['amount0Out']
                        df.loc[df['id'] == row['id'], 'amount1'] = row['amount1In']
                    else:
                        df.loc[df['id'] == row['id'], 'amount0'] = row['amount0In']
                        df.loc[df['id'] == row['id'], 'amount1'] = row['amount1Out']
            dfs.append(df)
        df.to_csv(f'{WD}insider/list/root_{dex}_list.csv', index=False)       
        
dfs = pd.concat(dfs)
dfs.drop(columns=['amount0In', 'amount0Out', 'amount1In', 'amount1Out'], inplace=True)
dfs
```
