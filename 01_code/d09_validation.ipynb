### Insider Trading

```python
dfi.loc[dfi['dex'] == 'uniswapv3_ethereum'].groupby('timestamp')['amountUSD'].sum().reset_index().set_index('timestamp').plot(rot=45)
```

```python
it = dfi.loc[dfi['dex'] == 'uniswapv3_ethereum'].groupby('timestamp')['amountUSD'].sum().reset_index()
it.rename(columns={'amountUSD':'insider_trading'}, inplace=True)

dex = 'uniswapv3_ethereum'
df = pd.read_csv(f'{WD}{dex}.csv', usecols = ['timestamp', 'amountUSD'])
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['timestamp'] = df['timestamp'].dt.strftime('%Y-%m-%d')
df.sort_values('timestamp', ascending=True, inplace=True)
tv = df.groupby('timestamp')['amountUSD'].sum().reset_index()
```

```python
df = tv.merge(it, on='timestamp', how='left', validate='1:1', indicator=True)
df['_merge'].value_counts()
df = df.loc[df['_merge'] == 'both', ['timestamp','amountUSD', 'insider_trading']]
df.set_index('timestamp')[['amountUSD', 'insider_trading']].plot(rot=45)
```
