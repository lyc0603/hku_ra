# Example 1: addresses more than 10000 (Ethereum/Binance)

- The Team Finance contract on Ethereum is 0xe2fe530c047f2d85298b07d9333c05737f1435fb.
- We can use Etherscan API to get a list of 'ERC20 - Token Transfer Events' by Address.
- Your need to sign up for a API key.
- However, the maximum number of data that we can get is 10000 and the number of token transfered event is more than 10000. Therefore, we need to paginate through blocks to get a full list of EOAs.

```python
dfs = []

for block in range(0, 100000000, 10000):
    
    url = f'https://api.etherscan.io/api\
?module=account\
&action=tokentx\
&address=0xe2fe530c047f2d85298b07d9333c05737f1435fb\
&page=1\
&offset=10000\
&startblock={block}\
&endblock={block+9999}\
&sort=asc\
&apikey=SE4Y8HN6P577WGIJSM8561XVS3279RCYAS' # Your API key on Etherscan
    
    r = requests.get(url)
    data = r.json()
    df = pd.DataFrame(data['result'])
    
    dfs.append(df)
    print(f'sucessfully get data on {block}/100000000.')
    
dfs = pd.concat(dfs)
dfs
```

```python
url = 'https://api.polygonscan.com/api\
?module=account\
&action=tokentx\
&address=0x3eF7442dF454bA6b7C1deEc8DdF29Cfb2d6e56c7\
&startblock=0\
&endblock=99999999\
&page=1\
&offset=10000\
&sort=asc\
&apikey=RQ81U42M4YSGZJ1DVTG8TCX64F446C4W8H' # Your API key on Polyscan.

r = requests.get(url)
r.status_code
data = r.json()
df = pd.DataFrame(data['result'])
df
```

## Data Processing

- From unix to datetime
- The amount of token transfered = value * 10^-tokenDecimal

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'], unit='s', origin='unix')
df['tokenAmount'] = df['value'].astype('float') / 10**df['tokenDecimal'].astype('float')
df
```
